***1. HTML Tag Removal***

```python
text1 = "<p>The<b>movie</b> <i>Inception</i> is a mind-bending thriller.</p>"
```

```python
import re

def removal_html_tags(text):
    pattern = re.compile('<.*?>')
    return pattern.sub('', text)
```

```python
removal_html_tags(text1)
```

```python
import re
def remove_url(text):
    pattern = re.compile(r'https?://\S+|www\.\S+')
    return pattern.sub(r'',text)
```

```python
remove_url(text1)
```

***Puntuation Removal***

```python
import string

exclude=string.punctuation
```

```python
def remove_punc(text):
    for char in exclude:
        text=text.replace(char,'')
    return text
```

```python
text2 = 'remove. with. Punctuation?'
```

```python
print(remove_punc(text2))
```
